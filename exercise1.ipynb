```python
#Q1
def string_ops(s):
    reversed_str = s[::-1]
    every_second = s[::2]
    return reversed_str, every_second

print(string_ops("hello world"))
```

```python
#Q2
def even_squares(nums):
    return [n**2 for n in nums if n % 2 == 0]

print(even_squares([1, 2, 3, 4, 5, 6]))
```

```python
#Q3
def char_frequency(s):
    freq = {}
    for ch in s:
        freq[ch] = freq.get(ch, 0) + 1
    return freq

print(char_frequency("banana"))
```

```python
#Q4
def odd_indices(nums):
    return [i for i, n in enumerate(nums) if n % 2 != 0]

print(odd_indices([1, 2, 3, 4, 5, 6]))
```

```python
#Q5
def rotate_right(lst, k):
    k = k % len(lst)
    return lst[-k:] + lst[:-k]

print(rotate_right([1, 2, 3, 4, 5], 2))
```

```python
#Q6
def lists_to_dict(keys, values):
    return dict(zip(keys, values))

print(lists_to_dict(["a", "b", "c"], [1, 2, 3]))
```

```python
#Q7
def manual_iterate(lst):
    it = iter(lst)
    while True:
        try:
            print(next(it))
        except StopIteration:
            break

manual_iterate([10, 20, 30])
```

```python
#Q8
def flatten(nested):
    return [item for sublist in nested for item in sublist]

print(flatten([[1, 2], [3, 4], [5]]))
```

```python
#Q9
def is_palindrome(s):
    return s == s[::-1]

print(is_palindrome("madam"))
print(is_palindrome("hello"))
```

```python
#Q10
def second_largest(nums):
    largest = second = float('-inf')
    for n in nums:
        if n > largest:
            second = largest
            largest = n
        elif largest > n > second:
            second = n
    return second

print(second_largest([4, 1, 7, 3, 7, 2]))
#haha
```
